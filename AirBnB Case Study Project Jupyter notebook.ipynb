```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb.groupby(['neighbourhood_group']).count()

# There are 5 neighbourhood groups and Manhattan shows up the most
```

```python
# Are private rooms the most popular in manhattan?

air_bnb.groupby(['neighbourhood_group', 'room_type']).count()


# In Manhattan private rooms are the second most common type of room and get the second most reviews per month
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb.groupby(['host_id']).count().sort_values(by = "number_of_reviews", inplace = False, ascending = False)

# Host 219517861 is the busiest with 327 reviews
```

```python
#Which neighorhood group has the highest average price?

air_bnb.groupby(['neighbourhood_group']).mean()

# Manhattan has the highest average price at 196.88 per night
```

```python
# Which neighbor hood group has the highest total price?

air_bnb.groupby(['neighbourhood_group']).sum()

# Manhattan had the highest total price at 4264527 toal
```

```python
#Which top 5 hosts have the highest total price?

air_bnb.groupby(['host_id']).sum().sort_values(by = "price", inplace = False, ascending = False).head(5)

#host_id's 219517861, 107434423, 156158778, 205031545, 30283594
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

host_bnb = air_bnb.groupby('host_id').sum()
zero_avail = host_bnb[host_bnb['availability_365'] == 0]
rev_100 = host_bnb[host_bnb['number_of_reviews'] >= 100]
booked_and_busy = zero_avail.merge(rev_100, on = 'host_id', how = 'inner')
booked_and_busy

#There are 136 hosts with no availability and a review count of 100 or more
```

```python
# What host has the highest total of prices and where are they located?

air_bnb.groupby(['host_id','neighbourhood_group']).sum().sort_values(by = "price", inplace = False, ascending = False)
# Host 219517861 who is located in Manhattan
```

```python
# When did Danielle from Queens last receive a review?

danielles = air_bnb[air_bnb['host_name']== 'Danielle']

d_from_q = danielles[danielles["neighbourhood_group"] == "Queens"]

d_from_q

# There are 8, all their last reviews were in 2019 or 2018, except 1 who has not been reviewed
```
